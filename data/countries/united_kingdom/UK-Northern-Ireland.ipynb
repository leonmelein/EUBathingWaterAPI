```python
import requests
import pandas as pd
import sqlite3
```

```python
server = "https://services-eu1.arcgis.com/kswen6BYexuc1SUk/arcgis/rest/services/BathingWatersDirectiveProtectedAreas/FeatureServer/0/query?where=1%3D1&outFields=Name,PA_CD,centroidX,centroidY&returnGeometry=false&outSR=4326&f=json"
```

```python
data = requests.get(server, headers={"Accept": "application/json, text/javascript, */*; q=0.01"})
```

```python
df = pd.json_normalize(data.json()['features'])
df = df.rename(columns={
    "attributes.PA_CD": "id",
    "attributes.Name": "name",
    "attributes.centroidX": "long",
    "attributes.centroidY": "lat"
})\
.set_index("id")\
# .drop(columns=[
#     'attributes.CLASS_DESCRIPTION', 'attributes.BW_URL',
#     'attributes.YEAR'
# ])
df['name'] = df['name'].str.title()
df['alternate_name'] = df['name']
locations = df[[
    'name', 'alternate_name', 'lat', 'long'
]].copy()
locations
```

```python
db = sqlite3.connect("../dataset.sqlite3")
locations.to_sql("locations", db, if_exists="append")
```
