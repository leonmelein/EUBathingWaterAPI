```python
import requests
import pandas as pd
import sqlite3
```

```python
server = "https://euswimapp.com/api/v1/beaches"
```

```python
data = requests.get(server, headers={"Accept": "application/json, text/javascript, */*; q=0.01"})
result = data.json()['beaches']
```

```python
df = pd.json_normalize(result)\
    .set_index('id')\
    .drop(columns=['address', 'image'])
df
```

```python
df.sort_values(by=['name'], ascending=True)
```
